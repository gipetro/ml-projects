```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
```

```python
matches18 = pd.read_csv('135_2018.csv', sep=';')
matches19 = pd.read_csv('135_2019.csv', sep=';')
matches20 = pd.read_csv('135_2020.csv', sep=';')
matches21 = pd.read_csv('135_2021.csv', sep=';')
matches22 = pd.read_csv('135_2022.csv', sep=';')
```

```python
matches18['match_date'] = pd.to_datetime(matches18['match_date'], utc=True)
matches18['target'] = (matches18['home_goals']-matches18['away_goals'] > 0).astype('int')

matches19['match_date'] = pd.to_datetime(matches19['match_date'], utc=True)
matches19['target'] = (matches19['home_goals']-matches19['away_goals'] > 0).astype('int')

matches20['match_date'] = pd.to_datetime(matches20['match_date'], utc=True)
matches20['target'] = (matches20['home_goals']-matches20['away_goals'] > 0).astype('int')

matches21['match_date'] = pd.to_datetime(matches21['match_date'], utc=True)
matches21['target'] = (matches21['home_goals']-matches21['away_goals'] > 0).astype('int')

matches22['match_date'] = pd.to_datetime(matches22['match_date'], utc=True)
matches22['target'] = (matches22['home_goals']-matches22['away_goals'] > 0).astype('int')
```

```python
rf22 = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
rf18 = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
rf19 = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
rf20 = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
rf21 = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train18 = matches18[matches18['match_date'] < '2019-02-01']
test18 = matches18[matches18['match_date'] > '2019-02-01']

train19 = matches19[matches19['match_date'] < '2020-02-01']
test19 = matches19[matches19['match_date'] > '2020-02-01']

train20 = matches20[matches20['match_date'] < '2021-02-01']
test20 = matches20[matches20['match_date'] > '2021-02-01']

train21 = matches21[matches21['match_date'] < '2022-02-01']
test21 = matches21[matches21['match_date'] > '2022-02-01']

train22 = matches22[matches22['match_date'] < '2022-11-01']
test22 = matches22[matches22['match_date'] > '2022-11-01']
```

```python
predictors = ['home_team_id', 'away_team_id', 'home_goals_for_avg_h', 'home_goals_against_avg_h', 'away_goals_for_avg_a', 'away_goals_against_avg_a']
```

```python
rf18.fit(train18[predictors], train18['target'])
rf19.fit(train19[predictors], train19['target'])
rf20.fit(train20[predictors], train20['target'])
rf21.fit(train21[predictors], train21['target'])
rf22.fit(train22[predictors], train22['target'])
```

```python
preds18 = rf18.predict(test18[predictors])
preds19 = rf19.predict(test19[predictors])
preds20 = rf20.predict(test20[predictors])
preds21 = rf21.predict(test21[predictors])
preds22 = rf22.predict(test22[predictors])
```

```python
acc18 = accuracy_score(test18['target'], preds18)
acc19 = accuracy_score(test19['target'], preds19)
acc20 = accuracy_score(test20['target'], preds20)
acc21 = accuracy_score(test21['target'], preds21)
acc22 = accuracy_score(test22['target'], preds22)

print("2018 accuracy score: ",acc18
      ,"\n2019 accuracy score: ",acc19
      ,"\n2020 accuracy score: ",acc20
      ,"\n2021 accuracy score: ",acc21,
      "\n2022 accuracy score: ",acc22)
```
