```python
import pandas as pd
raw_csv_data = pd.read_csv("Absenteeism-data.csv")
df = raw_csv_data.copy()
df = df.drop(['ID'], axis = 1)
df
```

```python
df_no_age = df.drop(['Age'], axis = 1)
df_no_age
```

```python
reasonColumns = pd.get_dummies(df['Reason for Absence'], dtype=int, drop_first = True)
```

```python
age_dummies = pd.get_dummies(df['Age'], dtype=int)
age_dummies
```

```python
df = df.drop(['Reason for Absence'], axis=1)
```

```python
reasonType1 = reasonColumns.loc[:, 1:14].max(axis=1)
reasonType2 = reasonColumns.loc[:, 15:17].max(axis=1)
reasonType3 = reasonColumns.loc[:, 18:21].max(axis=1)
reasonType4 = reasonColumns.loc[:, 22:].max(axis=1)
```

```python
reasonType4
```

```python
df = pd.concat([df, reasonType1, reasonType2, reasonType3, reasonType4], axis=1)
```

```python
df.columns.values
```

```python
columnNames = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason 1', 'Reason 2', 'Reason 3', 'Reason 4']
```

```python
df.columns = columnNames
```

```python
df_concatenated = pd.concat([df_no_age, age_dummies], axis = 1)
df_concatenated
```

```python
columnNamesReordered = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[columnNamesReordered]
```

```python
df_concatenated.columns.values
```

```python
concat_column_names_reorder = ['Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Daily Work Load Average', 'Body Mass Index',
       'Education', 'Children', 'Pets', 27,
       28, 29, 30, 31, 32, 33, 34, 36, 37, 38, 39, 40, 41, 43, 46, 47, 48,
       49, 50, 58, 'Absenteeism Time in Hours']
df_concatenated = df_concatenated[concat_column_names_reorder]
df_concatenated
```

```python
df_checkpoint = df.copy()
df_checkpoint
```

```python
df_checkpoint['Date'] = pd.to_datetime(df_checkpoint['Date'], format='%d/%m/%Y')
```

```python
monthsList = []
```

```python
for i in range(df_checkpoint.shape[0]):
    monthsList.append(df_checkpoint['Date'][i].month)
```

```python
df_checkpoint['Month Value'] = monthsList
```

```python
def dateToWeekday(dateVal):
    return dateVal.weekday()
```

```python
df_checkpoint['Day of the Week'] = df_checkpoint['Date'].apply(dateToWeekday)
```

```python
df_dateMod = df_checkpoint.drop(['Date'], axis = 1)
```

```python
column_names_reorder = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 'Month Value', 'Day of the Week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df_dateMod = df_dateMod[column_names_reorder]
df_dateMod
```

```python
dfDateMod = df_dateMod.copy()
```

```python
dfDateMod['Education'] = dfDateMod['Education'].map({1:0, 2:1, 3:1, 4:1})
```

```python
dfDateMod['Education'].value_counts()
```

```python
df_preprocessed = dfDateMod.copy()

df_preprocessed.to_csv('Absenteeism_preprocessedSelf.csv', index=False)
```
